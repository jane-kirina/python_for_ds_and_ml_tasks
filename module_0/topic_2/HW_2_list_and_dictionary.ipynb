**1. Списки - базові операції**

Створіть список температур за тиждень (у градусах Цельсія) з назвою
`temperatures` і значеннями [22, 25, 19, 23, 26, 20, 21, 22].


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть перший елемент
- Виведіть останній елемент
- Додайте температуру 24
- Видаліть температуру 19
- Виведіть довжину списку
- Обчисліть суму всіх температур
- Знайдіть максимальну температуру

```python
temperatures = [22, 25, 19, 23, 26, 20, 21, 22]
print(f'- Виведіть перший елемент: {temperatures[0]}')
print(f'- Виведіть останній елемент: {temperatures[-1]}')

print(f'- Додайте температуру 24:')
print(f'До додовання: {temperatures}')
temperatures.append(24)
print(f'Після додовання: {temperatures}')

print(f'- Видаліть температуру 19')
print(f'До видалення: {temperatures}')
temperatures.remove(19)
print(f'Після видалення: {temperatures}')

print(f'- Виведіть довжину списку: {len(temperatures)}')
print(f'- Обчисліть суму всіх температур: {sum(temperatures)}')
print(f'- Знайдіть максимальну температуру: {max(temperatures)}')
```


**2. Словники - базові операції**

Створіть словник про себе, який містить ключі `name`, `age`, `city` і значення, які відповідають вашему імені, віку і місту (можна вставити будь-які значення, не обовʼязково свої).


Виконайте (кожен пункт оцінюється в 0.5 бала):
- Виведіть значення ключа "name"
- Змініть значення ключа "age" на 21
- Додайте новий ключ "hobby" зі значенням "programming"
- Виведіть всі ключі
- Виведіть всі значення


```python
person = {
  'name': 'Jane',
  'age': 23,
  'city': 'Kharkiv'
}

print(f'- Виведіть значення ключа "name": {person['name']}')

print(f'- Змініть значення ключа "age" на 21')
print(f'До змін: {person['age']}')
person['age'] = 21
print(f'Після змін: {person['age']}')

print('- Додайте новий ключ "hobby" зі значенням "programming"')
print(f'До додовання нового ключа: {person.get('hobby')}')
person['hobby'] = 'programming'
print(f'Після додовання нового ключа: {person.get('hobby')}')

print(f'- Виведіть всі значення: {person.values()}')
```
