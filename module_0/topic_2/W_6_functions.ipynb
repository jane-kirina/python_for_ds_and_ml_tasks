**1. Функція без параметрів**

Створіть функцію `show_menu()`, яка приймає перелік страв, наприклад ['Піца', 'Паста', 'Салат'] і виводить меню ресторану в наступному вигляді:
```
=== МЕНЮ ===
1. Піца
2. Паста
3. Салат
=============
```

Викличте цю функцію на списку ["Суп грибний", "Котлета по-київськи", "Лимонад"].

````python
def show_menu():
  products =  ["Суп грибний", "Котлета по-київськи", "Лимонад"]
  start_of_menu = '=== МЕНЮ ==='
  end_of_menu  = '============='

  print(start_of_menu)
  for i, product in enumerate(products, start = 1):
    print(f'{i}. {product}')
  print(end_of_menu)

show_menu()
````

**2. Функція з параметром**

Створіть функцію `print_square(n)`, яка виводить квадрат числа.

Наприклад: `print_square(5)` → "Квадрат числа 5 = 25"

Викличте її з різними числами.

````python
def print_square(n):
  print(f'Квадрат числа {n} = {n ** 2}')

print_square(2)
print_square(10)
print_square(5)
print_square(8)
print_square(21)
````

**3. Функція з поверненням значення**

Створіть функцію `multiply(a, b)`, яка повертає добуток двох чисел.

Викличте функцію на власноруч заданих значеннях та виведіть результат.


````python
# Повна функція
def multiply(a, b):
  return a * b

print(f'Добутот чисіл 5 та 2: {multiply(5, 2)}')
print(f'Добутот чисіл 4 та 7: {multiply(4, 7)}')
print(f'Добутот чисіл 9 та 3: {multiply(9, 3)}')
print(f'Добутот чисіл 8 та 8: {multiply(8, 8)}')

````

````python
# З використанням lambda-функції
multiply = lambda x, y: x * y
print(f'Добутот чисіл 5 та 2: {multiply(5, 2)}')
print(f'Добутот чисіл 4 та 7: {multiply(4, 7)}')
print(f'Добутот чисіл 9 та 3: {multiply(9, 3)}')
print(f'Добутот чисіл 8 та 8: {multiply(8, 8)}')
````

**4. Функція пошуку максимумального значення**

Створіть функцію `find_max(numbers)`, яка:
- Приймає список чисел
- Повертає найбільше число в списку

Протестуйте на списку `[15, 42, 8, 23, 56, 19]`.


````python
# Зі встроєною функцією
# def find_max(numbers):
  # return max(numbers)

# З викоританням for-loop
def find_max(numbers):
  max_value = numbers[0]
  for num in numbers:
      if num > max_value:
          max_value = num
  return max_value

simple_list = [15, 42, 8, 23, 56, 19]
print(f'Найбільше число у списку {simple_list}: {find_max(simple_list)}')
````

**5. Функція з умовами**

Створіть функцію `check_grade(grade)`, яка:
- Приймає оцінку (число)
- Повертає "Здав" якщо grade >= 60
- Повертає "Не здав" якщо grade < 60

Протестуйте на різних оцінках.


````python
# Звичайна функція
def check_grade(grade):
  return "Здав" if grade >= 60 else "Не здав"

print(f'Оцінка 10 значить що учень \"{check_grade(10)}\"')
print(f'Оцінка 93 значить що учень \"{check_grade(93)}\"')
print(f'Оцінка 73 значить що учень \"{check_grade(73)}\"')
print(f'Оцінка 60 значить що учень \"{check_grade(60)}\"')
print(f'Оцінка 59 значить що учень \"{check_grade(59)}\"')
print(f'Оцінка 51 значить що учень \"{check_grade(51)}\"')
````

````python
# З використанням filter
def check_grade_with_filter(grade):
    result = list(filter(lambda x: x >= 60, [grade]))
    return "Здав" if result else "Не здав"

print(f'Оцінка 10 значить що учень \"{check_grade(10)}\"')
print(f'Оцінка 93 значить що учень \"{check_grade(93)}\"')
print(f'Оцінка 73 значить що учень \"{check_grade(73)}\"')
print(f'Оцінка 60 значить що учень \"{check_grade(60)}\"')
print(f'Оцінка 59 значить що учень \"{check_grade(59)}\"')
print(f'Оцінка 51 значить що учень \"{check_grade(51)}\"')
````
