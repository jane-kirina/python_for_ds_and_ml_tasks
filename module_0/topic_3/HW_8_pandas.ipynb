**5. Групування**

Обчисліть (кожен пункт оцінюється в 0.5 бала):
- Середній рахунок за статтю (groupby `sex`)
- Середні чайові за днем тижня (groupby `day`)
- Скільки відвідувачів у кожен день


```python
print('- Середній рахунок за статтю (groupby \'sex\'): ')
tip_df.groupby('sex')['total_bill'].mean()
```

```python
print('- Середні чайові за днем тижня (groupby \'day\'): ')
tip_df.groupby('day')['tip'].mean()
```

```python
print('- Скільки відвідувачів у кожен день: ')
print('Загальна кількість відвідувачів: ')
tip_df.groupby('day')['size'].sum()
```

```python
print('Кількість відвідувачів(не враховуючи кількість людей у компанії): ')
tip_df.groupby('day').size()
```
