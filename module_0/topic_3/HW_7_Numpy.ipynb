**1. Створення масивів**

Створіть (кожен пункт оцінюється в 0.5 бала):
- Масив зі списку `[5, 15, 25, 35, 45]`
- Масив з нулів розміром 7
- Масив з одиниць розміром 6
- Масив з числами від 10 до 19 використовуючи `np.arange()`

```python
import numpy as np

arr =  np.array([5, 15, 25, 35, 45])
print(f'Масив: {arr}')
arr_zeros = np.zeros(7)
print(f'Масив з нулями: {arr_zeros}')
arr_ones = np.ones(6)
print(f'Масив з одиницями: {arr_ones}')
arr_arange = np.arange(10, 20)
print(f'Масив з 10 до 19 включно: {arr_arange}')
```

**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
arr =  np.array([12, 24, 36, 48, 60])
print(f'Масив: {arr}')
print(f'- Помножте всі елементи на 2: {arr * 2}')
print(f'- Додайте до всіх елементів 10: {arr + 10}')
print(f'- Піднесіть всі елементи до квадрату: {arr ** 2}')
print(f'- Обчисліть квадратний корінь з усіх елементів: {np.sqrt(arr)}')
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

Обчисліть використовуючи функціонал numpy (кожен пункт оцінюється в 0.5 бала):
- Середнє значення
- Мінімум
- Максимум
- Суму всіх елементів

```python
data =  np.array([18, 27, 33, 41, 52, 68, 79])
print(f'Звичайний масив: {data}')
# З використанням np.function(array)
print(f'- Середнє значення: {np.mean(data)}')
print(f'- Мінімум: {np.min(data)}')
print(f'- Максимум: {np.max(data)}')
print(f'- Суму всіх елементів: {np.sum(data)}')
```

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr = np.array([15, 22, 35, 44, 58, 67])
print(f'Звичайний масив: {arr}')
print(f'- Перший елемент: {arr[0]}')
print(f'- Останній елемент: {arr[-1]}')
print(f'- Елементи з індексами 1 до 3: {arr[1:4]}')
print(f'- Елементи більші за 40: {arr[arr > 40]}')
```
