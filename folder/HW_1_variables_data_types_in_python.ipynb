**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4

var_sum = var1 + var2
var_diff = var1 - var2
var_production = var1 * var2
var_division = var1 / var2
var_integer_div = var1 // var2
var_remainder = var1 % var2
var_exponentiation = var1 ** var2

print(f'''
Початкові дані \'{var1}\' та \'{var2}\'. Операції:
\n Сумa: {var_sum}
\n Різниця: {var_diff}
\n Добуток: {var_production}
\n Ділення: {var_division}
\n Цілочисельне ділення: {var_integer_div}
\n Залишок від ділення: {var_remainder}
\n Піднесення до степеня: {var_exponentiation}
''')
```
